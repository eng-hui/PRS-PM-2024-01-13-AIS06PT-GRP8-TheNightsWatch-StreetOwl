```python
from ultralytics import YOLO
import os

project = os.path.join("..","..","runs")
```

```python
yolo_models = ["yolov8n","yolov8s", "yolov8m", "yolov8l", "yolov8x"] 
results_list = []
for model_name in yolo_models:
    model = YOLO(model_name)
    results = model.val(data="data.yaml", split="test",project=project,name=f"{model_name}_baseline_val")
    results_list.append(results)
    print(f"Model : {model_name}")
    print(results)
```

```python
all_results = {}

for i in range(0,5):
    print(yolo_models[i])
    print(results_list[i].results_dict)
    all_results[yolo_models[i]] = results_list[i].results_dict
    
```

```python
import pandas as pd

df = pd.DataFrame(all_results)
df.to_csv("basic_seg_model_val.csv")
df = df.style.background_gradient(cmap='magma', axis=1)

df
```
